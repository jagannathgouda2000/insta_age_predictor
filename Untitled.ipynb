```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train_age_dataset.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
df1=df[['Unnamed: 0','followers_avg_age','following_avg_age','max_repetitive_punc','number_of_words_per_action','content_views','weekends_trails_watched_per_day','slot1_trails_watched_per_day','slot2_trails_watched_per_day','slot3_trails_watched_per_day','slot4_trails_watched_per_day','avgt2','age_group']]
df1.head()
```

```python
y=df1.iloc[:,-1]
x=df1.iloc[:,0:-1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=14)
```

```python
knn.fit(x_train,y_train)
```

```python
pred = knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
accuracy_score(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
error_rate = []
# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(accuracy_score(y_test,pred_i))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
```
